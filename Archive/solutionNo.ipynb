```python
'''
Import data into data frame and display data
As data is delimited with a series of 0-valued rows, such rows are removed before further processing.
'''
df_src = pd.read_csv("badmintondata.csv")
df_clean = df_src.loc[~(df_src==0).all(axis=1)]
df_clean
```

```python
'''
Characteristics of data is printed
'''
print("Columns in data frame:\n======================================")
for i in df_clean.columns:
    print(i)
print("\n")
print("Unique values in each feature:\n======================================")
for i in range(0,5):
    print(df_clean.columns[i],":", df_clean[df_clean.columns[i]].unique())    
    
print("\n")
print("Outcome values in data frame:\n======================================")
for i in range(5,8):
    print(df_clean.columns[i])  
```

```python
'''
As the expected position of the shuttlecock in 3D space is a function of time, t.
A time feature is added according to the descriptor of the data set, where the data is collected every 10ms.
'''

idx = 0
new_index = []
new_timestamp = []
timestamp = 0

for i in range(len(df_clean.index)-1):
    if df_clean.index[i+1] == df_clean.index[i] + 1:
        new_index.append(idx)
        timestamp = timestamp + 10
        new_timestamp.append(timestamp)
    else:
        new_index.append(idx)
        timestamp = timestamp + 10
        new_timestamp.append(timestamp)
        idx = idx + 1
        timestamp = 0
        

new_index.append(idx)
new_timestamp.append(timestamp)


df_clean = df_clean.copy(deep=True)
df_clean['TESTSET'] = new_index
df_clean['TIMESTAMP'] = new_timestamp

'''
New list of columns are printed. Prep data frame is shown
'''
print("Columns in data frame:\n======================================")
for i in df_clean.columns:
    print(i)
print("\n")
df_clean


df_clean
df_clean.to_csv("check.csv")
```

```python
'''
Initial plot of data
'''

tmax = df_clean["TESTSET"].max()

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figwidth(50)
fig.set_figheight(10)

for i in range(tmax+1):
    # Generate some random data
    x = df_clean[df_clean["TESTSET"]==i]["TIMESTAMP"]
    y1 = df_clean[df_clean["TESTSET"]==i][df_clean.columns[7]]
    y2 = df_clean[df_clean["TESTSET"]==i][df_clean.columns[5]]-df_clean[df_clean["TESTSET"]==i]["HUMAN PLAYER POSITION (X) metres"]
    y3 = df_clean[df_clean["TESTSET"]==i][df_clean.columns[6]]-df_clean[df_clean["TESTSET"]==i]["HUMAN PLAYER POSITION (Y) metres"]

    # Plot the data
    ax1.scatter(x, y1, marker='x', s=np.ones_like(x)*2, alpha=0.8)
    ax2.scatter(x, y2, marker='x', s=np.ones_like(x)*2, alpha=0.8)
    ax3.scatter(x, y3, marker='x', s=np.ones_like(x)*2, alpha=0.8)

# Set labels and title
ax1.set_xlabel('Time')
ax1.set_ylabel(df_clean.columns[7])
ax1.set_title('Z position at time t')
ax2.set_xlabel('Time')
ax2.set_ylabel(df_clean.columns[5])
ax2.set_title('X position at time t w.r.f Human Starting X Position')
ax3.set_xlabel('Time')
ax3.set_ylabel(df_clean.columns[6])
ax3.set_title('Y position at time t w.r.f Human Starting Y Position')

# Show the plot
plt.show()
```

```python
'''
Splits data into Training and Testing set with a ratio of 70:30
'''

featureSet = df_clean.loc[:, ["HUMAN PLAYER POSITION (X) metres", "HUMAN PLAYER POSITION (Y) metres", "INITITAL VELOCITY OF SHUTTELCOCK(m/s)", "INITIAL SHUTTELCOCK FIRING ANGLE (DEGREE)", "SHUTTELCOCK SLANT ANGLE TO SIDELINE(DEGREE)", "TIMESTAMP"]]
outcomeSet = df_clean.loc[:, ["SHUTTLECOCK POSITIION IN AIR(X ) metres","SHUTTLECOCK POSITIION IN AIR(Y) metres","SHUTTLECOCK POSITIION IN AIR(Z) metres"]]

XTrain, XTest, YTrain, YTest = train_test_split(featureSet, outcomeSet, test_size=0.3, random_state=333)
```

```python
X = XTrain

YHeight = YTrain["SHUTTLECOCK POSITIION IN AIR(Z) metres"].to_list()
YSide = YTrain["SHUTTLECOCK POSITIION IN AIR(X ) metres"].to_list()
YForward = YTrain["SHUTTLECOCK POSITIION IN AIR(Y) metres"].to_list()

poly = PolynomialFeatures(3)
polyFeatures = poly.fit_transform(X)

regZ = LinearRegression()
regZ.fit(polyFeatures,YHeight)
regY = LinearRegression()
regY.fit(polyFeatures,YForward)
regX = LinearRegression()
regX.fit(polyFeatures, YSide)


print("Z predictor R-score on Train: ",regZ.score(poly.fit_transform(XTrain),YTrain["SHUTTLECOCK POSITIION IN AIR(Z) metres"]))
print("Z predictor R-score on Test: ",regZ.score(poly.fit_transform(XTest),YTest["SHUTTLECOCK POSITIION IN AIR(Z) metres"]))

print("Y predictor R-score on Train: ",regY.score(poly.fit_transform(XTrain),YTrain["SHUTTLECOCK POSITIION IN AIR(Y) metres"]))
print("Y predictor R-score on Test: ",regY.score(poly.fit_transform(XTest),YTest["SHUTTLECOCK POSITIION IN AIR(Y) metres"]))

print("X predictor R-score on Train: ",regX.score(poly.fit_transform(XTrain),YTrain["SHUTTLECOCK POSITIION IN AIR(X ) metres"]))
print("X predictor R-score on Test: ",regX.score(poly.fit_transform(XTest),YTest["SHUTTLECOCK POSITIION IN AIR(X ) metres"]))

#use model to make predictions on response variable
yZ_predicted = regZ.predict(poly.fit_transform(XTest))
yY_predicted = regY.predict(poly.fit_transform(XTest))
yX_predicted = regX.predict(poly.fit_transform(XTest))


fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax3.remove()

ax3 = fig.add_subplot(1,3,3,projection='3d')
#create scatterplot of x vs. y
ax1.scatter(XTest["TIMESTAMP"], YTest["SHUTTLECOCK POSITIION IN AIR(Z) metres"], s=np.ones_like(yZ_predicted), alpha=0.1)
ax1.scatter(XTest["TIMESTAMP"], yZ_predicted, c='r', s=np.ones_like(yZ_predicted), alpha=0.1)

ax2.scatter(XTest["TIMESTAMP"], YTest["SHUTTLECOCK POSITIION IN AIR(X ) metres"], s=np.ones_like(yX_predicted), alpha=0.1)
ax2.scatter(XTest["TIMESTAMP"], yX_predicted, c='r', s=np.ones_like(yX_predicted), alpha=0.1)

ax3.scatter3D(XTest["TIMESTAMP"],XTest["SHUTTELCOCK SLANT ANGLE TO SIDELINE(DEGREE)"],YTest["SHUTTLECOCK POSITIION IN AIR(Y) metres"], marker=".", s=np.ones_like(XTest["TIMESTAMP"]))
ax3.scatter3D(XTest["TIMESTAMP"],XTest["SHUTTELCOCK SLANT ANGLE TO SIDELINE(DEGREE)"] ,yY_predicted, color='red', s=np.ones_like(XTest["TIMESTAMP"]), alpha=0.1)

# Set labels and title
ax1.set_xlabel('Time')
ax1.set_ylabel(df_clean.columns[7])
ax1.set_title('Z position at time t')
ax2.set_xlabel('Time')
ax2.set_ylabel(df_clean.columns[5])
ax2.set_title('X position at time t')
ax3.set_xlabel('Time')
ax3.set_ylabel(df_clean.columns[4])
ax3.set_zlabel(df_clean.columns[6])
ax3.set_title('Y position at time t')


fig.set_figwidth(20)
```
